```python
import math

def calculate_angle(H, M):
    # Ensure H is within 0-11 for calculation (12 becomes 0)
    H = H % 12

    # Calculate the angle for the minute hand (6 degrees per minute)
    minute_angle = M * 6

    # Calculate the angle for the hour hand
    # (30 degrees per hour + 0.5 degrees per minute)
    hour_angle = (H * 30) + (M * 0.5)

    # Calculate the absolute difference between the angles
    angle_diff = abs(hour_angle - minute_angle)

    # The minimum angle can be the difference or 360 - difference
    min_angle = min(angle_diff, 360 - angle_diff)

    # Return the floor of the minimum angle
    return math.floor(min_angle)

# Test with the provided example:
H = 9
M = 0
result = calculate_angle(H, M)
print(result)

# Another test case (e.g., 3:30)
H_test = 3
M_test = 30
result_test = calculate_angle(H_test, M_test)
print(f"The angle at {H_test}:{M_test} is: {result_test}")
```

```python
# Program to calculate in-hand monthly salary after deductions and tax

# Take annual CTC as input in lakhs
ctc_input_lakhs = float(input("Enter your annual CTC in Lakhs: "))
ctc_annual = ctc_input_lakhs * 100000  # Convert lakhs to actual amount

# Calculate deductions (HRA, DA, PF) based on CTC
hra_deduction_amount = ctc_annual * 0.10  # 10% HRA
da_deduction_amount = ctc_annual * 0.05   # 5% DA
pf_deduction_amount = ctc_annual * 0.03   # 3% PF

# Calculate the income after HRA, DA, PF deductions, which is then subject to tax
taxable_income = ctc_annual - hra_deduction_amount - da_deduction_amount - pf_deduction_amount

# Convert taxable income to lakhs for slab comparison
taxable_income_lakhs = taxable_income / 100000

# Calculate tax based on the given slabs (progressive taxation)
tax_amount = 0

if taxable_income_lakhs < 5:
    tax_amount = 0
elif taxable_income_lakhs < 10:  # 5-10 lakhs slab: 10% tax on amount above 5 lakhs
    tax_amount = (taxable_income_lakhs - 5) * 0.10 * 100000
elif taxable_income_lakhs < 20:  # 10-20 lakhs slab: 10% on 5 lakhs + 20% on amount above 10 lakhs
    tax_amount = (5 * 0.10 * 100000) + ((taxable_income_lakhs - 10) * 0.20 * 100000)
else:  # Above 20 lakhs slab: 10% on 5 lakhs + 20% on 10 lakhs + 30% on amount above 20 lakhs
    tax_amount = (5 * 0.10 * 100000) + (10 * 0.20 * 100000) + ((taxable_income_lakhs - 20) * 0.30 * 100000)

# Calculate net annual salary after tax
net_annual_salary = taxable_income - tax_amount

# Calculate in-hand monthly salary
in_hand_monthly_salary = net_annual_salary / 12

# Display the results
print(f"\n--- Salary Calculation Summary ---")
print(f"Annual CTC: {ctc_annual:,.2f} INR")
print(f"HRA Deduction (10%): {hra_deduction_amount:,.2f} INR")
print(f"DA Deduction (5%): {da_deduction_amount:,.2f} INR")
print(f"PF Deduction (3%): {pf_deduction_amount:,.2f} INR")
print(f"\nTaxable Income: {taxable_income:,.2f} INR")
print(f"Annual Tax Amount: {tax_amount:,.2f} INR")
print(f"\nNet Annual Salary: {net_annual_salary:,.2f} INR")
print(f"In-hand Monthly Salary: {in_hand_monthly_salary:,.2f} INR")
```

```python
ctc = float(input("Enter your annual CTC in lakhs: "))
ctc_annual = ctc * 100000
hra_ctc = ctc_annual * 0.1
da_ctc = ctc_annual * 0.05
pf_ctc = ctc_annual * 0.03
ctc_actual = ctc_annual - hra_ctc - da_ctc - pf_ctc

taxable_income_lakhs = ctc_actual / 100000
tax_amount = 0

if taxable_income_lakhs < 5:
    tax_amount = 0
elif taxable_income_lakhs < 10:
    # Tax on income above 5 lakhs at 10%
    tax_amount = (taxable_income_lakhs - 5) * 0.10 * 100000
elif taxable_income_lakhs < 20:
    # Tax on 5-10 lakh slab (5 lakhs at 10%) + tax on income above 10 lakhs at 20%
    tax_amount = (5 * 0.10 * 100000) + ((taxable_income_lakhs - 10) * 0.20 * 100000)
else: # Handles taxable_income_lakhs >= 20
    # Tax on 5-10 lakh slab (5 lakhs at 10%) + tax on 10-20 lakh slab (10 lakhs at 20%) + tax on income above 20 lakhs at 30%
    tax_amount = (5 * 0.10 * 100000) + (10 * 0.20 * 100000) + ((taxable_income_lakhs - 20) * 0.30 * 100000)

net_annual_salary = ctc_actual - tax_amount
in_hand_monthly_salary = net_annual_salary / 12

print(f"Annual CTC: {ctc_annual:,.2f} INR")
print(f"HRA Deduction (10%): {hra_ctc:,.2f} INR")
print(f"DA Deduction (5%): {da_ctc:,.2f} INR")
print(f"PF Deduction (3%): {pf_ctc:,.2f} INR")
print(f"Taxable Income: {ctc_actual:,.2f} INR (or {taxable_income_lakhs:,.2f} Lakhs)")
print(f"Annual Tax Amount: {tax_amount:,.2f} INR")
print(f"Net Annual Salary: {net_annual_salary:,.2f} INR")
print(f"In-hand Monthly Salary: {in_hand_monthly_salary:,.2f} INR")
```
